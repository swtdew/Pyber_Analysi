```python
# Import dependencies
import matplotlib.pyplot as plt
```

**NOTE**: Chaining .pyplot to matplotlib is the same as typing "from matplotlib import pyplot". Use the alias 'plt' for pyplot. The pyplot module allow us to create all the necessary charts; create a plotting area in a figure; plot lines in a plotting area; and annotate the chart with labels, colors, and other decorative features. Pyplot provides a MATLAB-like interface for making plots.

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# Create a Line Chart using the MATLAB Method

```python
# Create the plot with plt.plot() --- 'plt.plot()' is the MATLAB method
plt.plot(x_axis, y_axis)
```

# Create a Line Chart using the Object-Oriented Interface Method

```python
# Create the plot with ax.plt() --- 'ax.plot()' is the object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

**NOTE**: The output is the same as using the MATLAB method.

#### Another way to create a line chart using the object-oriented interface method:

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

**NOTE**: The result is the same as using the MATLAB method.

#### Create a quick graph using object-oriented interface method:

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

# Create a Scatter Plot using the MATLAB Method
Scatter plots have many benefits over line and bar charts. A scatter plot visually depicts a positive, negative, or no-correlation relationship between two factors.

In the example scatter plot, we determine that there is no correlation between the number of rides and the average fare, because the r value is positive and close to zero. This is called a **regression analysis**.



### There are two ways to create scatter plots using the MATLAB Method:
1. plt.plot()
2. plt.scatter()

```python
# Create a scatter plot with the plt.plot() function
plt.plot(x_axis, y_axis, 'o')
```

**NOTE**: When we use the plt.plot() function to create a scatter plot, we need to add a lowercase "o" as a parameter inside the parantheses. This switches the plot from a line chart to a scatter plot.

```python
# Create a scatter plot with the plt.scatter() function
plt.scatter(x_axis, y_axis)
```

**NOTE**: The output using plt.scatter(x_axis, y_axis) is the same as using plt.plot(x_axis, y_axis, 'o').

# Create a Bubble Chart using the MATLAB Method

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

**NOTE**: To change the marker size, use the s, or size parameter, for the marker. The s parameter must be equal to a feature of our data. Here, we are making the size of each marker equal to the data floating-point decimal values of the y-axis by using s=y_axis.

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

**NOTE**: We can adjust the size of the markers by multiplying the data in the y-axis by any value. Here, we are iterating through the y-axis and multiplying each data point by 3 and adding it to a new list.

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

**NOTE**: We can refactor our code so that we don't use a for loop to create the new list (y_axis_larger) and use Python's **list comprehension** instead. You can use a list comprehension to replace many for and while loops. The format for list comprehension: new_list = [expression for item in list if conditional]

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

**NOTE**: The output is the same as before.

#### Skill Drill in 5.1.7: Create a Bubble Chart using object-oriented method
* Change the color of the markers to sky blue.
* Change the size of the markers to 5 times each data point.
* Make the color 20% transparent.
* Add a black edge color to the circles.
* Make the linewidth of the circles 2 points.
* Add a legend for the city of Boston.
* Add a title.
* Switch the axis so that the Fare data is on the x-axis.

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, 
           x_axis, 
           fc='#75bbfd', 
           alpha= 0.2, 
           edgecolor="black", 
           linewidth=2, 
           s = [i * 5 for i in y_axis], 
           label="Boston")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.legend()
```

# Create a Pie Chart using the MATLAB Method

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

We explode on the largest percentage, which is the 7th value, which is July. We use array-like data points like the tuple explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0). The length of the tuple should be equal to the number of wedges in the pie chart. To pop out the month of July, we use a decimal-point value of 0.2. To add percentage in each wedge, we use the autopct paremeter and provide the format of one decimal using .1f%. The % before and after the .1f$ formats the number as a percentage.

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

To fix the crowding of the pie chart, increase the size of the pie chart using the first line of code before the rest.

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", 
          "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", 
          "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')
plt.show()
```

#### Skill Drill in 5.1.8: Create a pie chart using the object-oriented interface method
* Add a percentage to one decimal place to each wedge of the pie.
* Increase the figure size to 8x8.
* Explode the two highest percentage months.
* Add a shadow.
* Add a start angle so that January is at the top.
* Reverse the order so that the month order is in a clockwise direction.
* Add new colors of your own choosing or use the colors from the previous pie chart

```python
fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", 
          "yellowgreen", "greenyellow", "yellow", "orange", 
          "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       shadow=True,
       startangle=90,
       counterclock=False,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```
