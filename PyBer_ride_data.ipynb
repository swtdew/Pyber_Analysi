# Plot data directly using a DataFrame or DataSeries
Use the plot() function by adding the series (ds) or DataFrame (df) in the following format: ds.plot() or df.plot()

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

Adjust the x-ticks to show all the months by adjusting the code:

```python
# Set x-axis and tick locations
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

First, we need an array that is the length of the DataFrame. Using the NumPy module, we create an array of numbers using the arange function for the length of the pyber_ride_df. Then, we set the tick locations for every value (in this case, it's every month), in the x-axis using a list comprehension. Lastly, we add the xticks() function and pass the tick_locations and labels of the x-ticks, pyber_ride_df["Month"]

# Create a bar chart with the same data
### There are two ways to do this...
In the 1st method, we chain the bar() function to the plot() function:

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

In the 2nd method, we add the kind parameter to the plot() function:

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
```

#### Skill Drill in 5.1.10: Using the data from the PyBer ride DataFrame, create a bar chart with the following annotations and characteristics:
* Add error bars.
* Add caps to the error bars.
* Change the color of the bars to sky blue.
* Rotate the labels on the x-axis to horizontal.
* Set the y-axis increment to every $5.

```python
# 1. Add error bars
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])

pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", yerr=stdev)
plt.show()
```

```python
#2. Add caps to the error bars
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=3)
plt.show()
```

```python
# 3. Change the color of the bars to sky blue
pyber_ride_df.plot.bar(x="Month", 
                       y="Avg. Fare ($USD)", 
                       yerr=stdev, 
                       capsize=3, 
                       color="#75bbfd")
plt.show()
```

```python
#4. Rotate the labels on the x-axis to horizontal
pyber_ride_df.plot.bar(x="Month", 
                       y="Avg. Fare ($USD)", 
                       yerr=stdev, 
                       capsize=3, 
                       color="#75bbfd")

plt.xticks(rotation=0)

plt.show()
```

```python
#5. Set the y-axis increment to every $5
pyber_ride_df.plot.bar(x="Month", 
                       y="Avg. Fare ($USD)", 
                       yerr=stdev, 
                       capsize=3, 
                       color="#75bbfd")

plt.xticks(rotation=0)

plt.yticks(np.arange(0, 55, step=5.0))

plt.show()
```
