```python
# Import dependencies
import matplotlib.pyplot as plt
import statistics
```

**NOTE**: We are importing a new dependency, **the statistics module**, which comes with the Anaconda installation. Weâ€™re going to use the statistics module so that we can add the y-error bars to line and bar charts. With the statistics module, we can calculate the standard deviation of the fare amount.

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

# Add Error Bars to a Line Chart
Adding error bars can show either a standard deviation, standard error, confidence intervals, or minimum or maximum values of a dataset.

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

Make the error bars look more professional by adding a "cap" to the error bars that's equal to a scalar value. To do this, we use the capsize= parameter.

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

# Add Error Bars and a Capsize using the Object-Oriented Interface Method

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

**NOTE**: The output is the same graph as using the MATLAB method.

# Adding Error Bars and a Capsize to a Bar Chart using the MATLAB Method

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

# Adding Error Bars and a Capsize to a Bar Chart using the Object-Oriented Interface Method

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

**NOTE**: The output looks like the same graph as using the MATLAB method.

# Change the Major Ticks

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

**NOTE**: Compare this horizontal bar graph to the first one. The major ticks are spaced by 5 dollar increment, instead of 10 dollar increment, to better estimate the averages fares. To create the $5 increment for the x-ticks, we use the numpy module to create an array of numbers using the arange() function. We defined the lower and upper limit of the x-axis and the increment frequency with the code: np.arange(0, 51, step=5.0)

```python
import numpy as np
np.arange(0, 51, step=5.0)
```

```python
np.arange(0, 51)
```

### Create the same graph using the Object-Oriented Interface Method

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

Alternatively, you can also use ax.xaxis.set_ticks(). If the y-axis had numerical values, you could change how often every major y-tick occurs by using the yticks() function for the MATLAB method and the set_yticks() function for the Object-Oriented Interface Method.

# Add Minor Ticks
To add minor tickets to the x-axis on horizontal bar chart, use the set_minor_locator function with the xaxis class in the object-oriented interface method. Inside the set_minor_locator functino, add the MultipleLocator method and pass how often you want the minor ticks.

```python
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure
fig, ax = plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))

plt.show()
```

Now we have minor tickers every 1 dollar increment, and major ticks at every 5 dollar increment.
